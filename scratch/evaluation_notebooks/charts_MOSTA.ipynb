```python
import pandas as pd
import matplotlib.pyplot as plt


# Set global font sizes
plt.rcParams.update({
    'font.size': 20,          # Default font size
    'axes.titlesize': 20,     # Title font size
    'axes.labelsize': 20,     # x and y label font size
    'xtick.labelsize': 20,    # x tick font size
    'ytick.labelsize': 20,    # y tick font size
    'legend.fontsize': 20,    # Legend font size
    'figure.titlesize': 20    # Figure title font size
})
```

### MOSTA

```python
# Create the data


# Seventh dataset
data_ivp_ti6 = {
    "variant": [
        "EOT",
        "PACM g+mc",
        "PACM g+lr",
        "PACM g+mc+lr",
        "PROT+mc",
        "PROT+lr",
        "PROT+mc+lr"
    ],
    "W. Wasserstein": [3.25110, 3.13688, 2.93821, 3.20032, 3.24435, 5.20044, 2.81416],
    "Wasserstein": [3.41809, 4.09322, 3.90387, 3.55521, 3.94641, 6.30637, 3.23308],
    "MMD": [0.08968, 0.08560, 0.08805, 0.08404, 0.08909, 0.12258, 0.09255],
    "Energy": [9.22648, 11.94767, 9.86420, 9.26921, 8.79733, 45.86176, 10.31876]
}


data_next_step_ti6 = {
    "variant": [
        "EOT",
        "PACM g+mc",
        "PACM g+lr",
        "PACM g+mc+lr",
        "PROT+mc",
        "PROT+lr",
        "PROT+mc+lr"
    ],
    "W. Wasserstein": [1.89199, 1.86546, 1.88158, 1.88780, 1.88370, 1.89782, 1.87102],
    "Wasserstein": [1.87348, 1.85180, 1.86900, 1.83861, 1.86244, 1.86646, 1.91935],
    "MMD": [0.16369, 0.15922, 0.16081, 0.16099, 0.16350, 0.16690, 0.16377],
    "Energy": [11.61465, 11.48248, 11.39942, 11.47519, 11.49878, 11.79514, 11.63877]
}


```

```python
settings= [
        "EOT",
        # "PACM g+mc",
        "PACM g+lr",
        "PACM g+mc+lr",
        # "PROT+mc",
        "PROT+lr",
        "PROT+mc+lr"
    ]
# Convert to DataFrame
df_ivp6 = pd.DataFrame(data_ivp_ti6)
df_next_step6 = pd.DataFrame(data_next_step_ti6)

df_ivp6 = df_ivp6 [df_ivp6["variant"].isin(settings)]
df_next_step6 = df_next_step6 [df_next_step6["variant"].isin(settings)]
```

## Wassersteins

```python

# # Plot 1
plt.figure(figsize=(12, 6))
# metrics = ["Wasserstein"]
metrics = ["W. Wasserstein"]

for metric in metrics:
    plt.plot(df_ivp6["variant"], df_ivp6[metric], marker="o", label=metric+"6")

plt.title("IVP Evaluation")
plt.xlabel("Variant")
plt.ylabel("Metric Value")
plt.legend()
plt.grid(True)
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()


# # Plot 2
plt.figure(figsize=(12, 6))
# metrics = ["Wasserstein"]
metrics = ["W. Wasserstein"]

for metric in metrics:
    plt.plot(df_next_step6["variant"], df_next_step6[metric], marker="o", label=metric+"timestep: 6")

plt.title("Next Step Evaluation")
plt.xlabel("Variant")
plt.ylabel("Metric Value")
plt.legend()
plt.grid(True)
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()



# # Plot 3
plt.figure(figsize=(12, 6))
# metrics = ["Wasserstein"]
metrics = ["W. Wasserstein"]

for metric in metrics:
    plt.plot(df_ivp6["variant"], df_ivp6[metric], marker="o", label='IVP: '+metric)

    plt.plot(df_next_step6["variant"], df_next_step6[metric], marker="o", label='Next Step: '+metric)

plt.title("IVP and Next Step Evaluation")
plt.xlabel("Variant")
plt.ylabel("Metric Value")
plt.legend()
plt.grid(True)
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()


# # Plot 4
def rel_to_baseline(s): 
    base = s.iloc[0]
    return 100*(s/base - 1.0)

plt.figure(figsize=(12,6))
plt.plot(df_ivp6['variant'], rel_to_baseline(df_ivp6['Wasserstein']), marker='o', label='IVP: Wasserstein')
plt.plot(df_next_step6['variant'], rel_to_baseline(df_next_step6['Wasserstein']), marker='s', label='Next Step: Wasserstein')

plt.axhline(0, color='gray', lw=1, alpha=0.5)
plt.ylabel('% change vs M-OT'); plt.xlabel('Variant'); plt.legend(); plt.grid(True); plt.xticks(rotation=20)
plt.title("Relative improvement of Variants (Wasserstein)")
plt.tight_layout(); plt.show()


# # Plot 5
def rel_to_baseline(s): 
    base = s.iloc[0]
    return 100*(s/base - 1.0)

plt.figure(figsize=(12,6))
plt.plot(df_ivp6['variant'], rel_to_baseline(df_ivp6['W. Wasserstein']), marker='o', label='IVP: W. Wasserstein')
plt.plot(df_next_step6['variant'], rel_to_baseline(df_next_step6['W. Wasserstein']), marker='s', label='Next Step: W. Wasserstein')

plt.axhline(0, color='gray', lw=1, alpha=0.5)
plt.ylabel('% change vs M-OT'); plt.xlabel('Variant'); plt.legend(); plt.grid(True); plt.xticks(rotation=20)
plt.title("Relative improvement of Variants (W. Wasserstein)")
plt.tight_layout(); plt.show()

```
